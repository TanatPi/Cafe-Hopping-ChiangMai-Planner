```python
import osmnx as ox
import networkx as nx
import pandas as pd
from ast import literal_eval
import numpy as np
import random
import matplotlib.colors as mcolors
import colorsys
ox.config(log_console=True,
          use_cache=True)
```

```python
mcolors.CSS4_COLORS.values()
```

```python
def hex_to_hsv(hex_color):
    # Convert hex to RGB
    r = int(hex_color[1:3], 16) / 255.0
    g = int(hex_color[3:5], 16) / 255.0
    b = int(hex_color[5:7], 16) / 255.0

    # Convert RGB to HSV
    hsv = colorsys.rgb_to_hsv(r, g, b)

    return hsv

def filter_colors_in_dict(color_dict, min_saturation=0.8, min_value=0, max_saturation = 1.0, max_value = 0.5):
    filtered_colors = {}

    for key, hex_color in color_dict.items():
        hsv = hex_to_hsv(hex_color)
        saturation, value = hsv[1], hsv[2]

        if saturation > min_saturation and value > min_value and saturation < max_saturation and value < max_value:
            filtered_colors[key] = hex_color

    return filtered_colors
```

```python
def generate_random_color_names(n):
    # Get a list of named colors from matplotlib
    all_colors = mcolors.CSS4_COLORS

    selceted_colors = list(filter_colors_in_dict(all_colors, min_saturation=0.5, min_value=0, max_saturation = 1.0, max_value = 0.9).keys())

    # Shuffle the list to get random colors
    random.shuffle(selceted_colors)

    # Select the first n colors
    random_color_names = selceted_colors[:n]

    return random_color_names
```

```python
d = pd.read_csv('data.csv')
s = pd.read_csv('E:\Work\DS\Optimization\Term Project\git\output\solution_to_plot\solutions.csv')
```

```python
Point = d['lat_long'].iloc[s.values.T[0]]
Point = np.array([literal_eval(coord) for coord in Point])
# Point
```

```python
G = ox.graph_from_point((18.78819354025208, 98.98622549572866), dist=3000, network_type='drive')
```

```python
Node = []
for i in Point:
    Node.append(ox.nearest_nodes(G, i[1], i[0]))

r = []
for i in range(len(Node)-1):
    r.append(nx.shortest_path(G, Node[i], Node[i+1], weight='length'))
```

```python
n_of_location = len(r)
colors = generate_random_color_names(n_of_location)
m1 = ox.plot_route_folium(G, r[0], popup_attribute='length', weight=7, color=colors[0])
for i in range(1,len(r)):
    m1 = ox.plot_route_folium(G, r[i],route_map=m1, popup_attribute='length', weight=7, color=colors[i])
m1
```
